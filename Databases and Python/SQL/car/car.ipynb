# Task
1. Take a dataset from attached URL https://archive.ics.uci.edu/ml/datasets/Car+Evaluation
2. Create a database called as cardataset
3. Create a table called as car with a column name given in dataset description
4. Dump all the data into car table
5. Try to check whether all the data is available inside your table or not
6. Try to group all the data with COL1 AND count occurences of each and every record based on col1 value
7. Try to filter a record where col 3 value will be 4
8. Try to update a col 3 value with 8 wherever you have value equal to 2
9. Try to delete table
10. Try to delete database

```python
import mysql.connector
import csv

mydb = mysql.connector.connect(
    host='localhost',
    user='abc',
    password='password'
)

cursor = mydb.cursor()

# Creating a database called cardataset
cursor.execute('CREATE DATABASE IF NOT EXISTS cardataset')

# Creating a table called car
cursor.execute('''CREATE TABLE IF NOT EXISTS cardataset.car(
    buying VARCHAR(10),
    maint VARCHAR(10),
    doors VARCHAR(10),
    persons VARCHAR(10),
    lug_boot VARCHAR(10),
    saftey VARCHAR(10),
    Class VARCHAR(10))''')

# Dumping the data into car table
with open('car.data') as file:
    reader = csv.reader(file)

    for record in reader:
        cursor.execute(f'INSERT INTO cardataset.car VALUES {tuple(record)}')
    mydb.commit()

# Check if the data is available inside the car table
cursor.execute('SELECT * FROM cardataset.car')
cursor.fetchall()
```

```python
# Try to group all the data with col1 and and count occurences of each and every record based on col1 value
cursor.execute('SELECT buying, COUNT(*) FROM cardataset.car GROUP BY buying')
cursor.fetchall()
```

```python
# Try to filter a record where col3 value will be 4
cursor.execute('SELECT * FROM cardataset.car WHERE doors = "4"')
cursor.fetchall()
```

```python
# Try to update a col3 value with 8 wherever you have value equal to 2
cursor.execute('UPDATE cardataset.car SET doors = "8" WHERE doors = "2"')
mydb.commit()
```

```python
# To see the updation
cursor.execute('SELECT * FROM cardataset.car')
cursor.fetchall()
```

```python
# Try to delete table
cursor.execute('DROP TABLE cardataset.car')
```

```python
# Try to delete database
cursor.execute('DROP DATABASE cardataset')
```
