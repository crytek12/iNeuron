##### Task
Try Create new columns based on the keys which is available a user dictionary, which is a part of dictionary (0, 1, 2, 3, ... till 30)

```python
import pandas as pd
import requests


url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
```

```python
data = requests.get(url)
data1 = data.json()
```

```python
df = pd.DataFrame(data1, columns=['id', 'url', 'repository_url', 'labels_url', 'user'])
```

```python
for index, dic in enumerate(df['user']):
    for key, value in dic.items():
        if type(value) != int:
            df.loc[index, key] = value
        else:
            df.loc[index, key] = int(value)
```
